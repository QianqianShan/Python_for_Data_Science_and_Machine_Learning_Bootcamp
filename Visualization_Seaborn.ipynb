## Seaborn

Seaborn is designed to work well with pandas DataFrame objects. See more instructions from http://seaborn.pydata.org/

### Distribution plots

```python
!python --version
```

```python
import seaborn as sns 
import pandas as pd
%matplotlib inline 
```

```python
tips = pd.read_csv('data/tips.csv')
tips.head()
```

```python
sns.distplot(tips['total_bill'])
```

```python
# remove kernel density estimate 
sns.distplot(tips['total_bill'], kde = False)
```

```python
# play around with bins (number of 'buckets')
sns.distplot(tips['total_bill'], kde = False, bins = 40)
```

### Joint plots

Two distplots on one plot.

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips)
```

```python
# use kind to specify the scatter plot kind
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'hex') # hexagon
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'reg') # regression
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'kde') # kde
```

### Pairplot

Pairwise relationships across the entire dataframe.

* jointplot of every single possible combination of numerical colums in data frame.

```python
sns.pairplot(tips)
```

```python
# add hue for categorical columns 
sns.pairplot(tips, hue = 'sex')
```

```python
# use palette to adjust the colors
sns.pairplot(tips, hue = 'sex', palette = 'coolwarm')
```

### rugplot

Draw a dashmark for every point along the distribution line.

```python
sns.rugplot(tips['total_bill'])
```
